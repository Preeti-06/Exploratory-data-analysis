## Exploratory Data Analysis

### Gender

```python
s = sns.countplot(x = 'Gender', data = df,palette = 'deep')
for bars in s.containers:
     s.bar_label(bars)
```

```python
cui = df.groupby(['Gender'],as_index =False)['Overall Rating'].sum().sort_values(by = 'Overall Rating',ascending = False)
sns.barplot(data = cui,x = 'Gender',y = 'Overall Rating',palette = 'deep')
```

### from above graph we conclude that their is maximum number of male who rate and least number of females

### Smoker

```python
s = sns.countplot(x = 'Smoker', data = df,palette = 'deep')
for bars in s.containers:
     s.bar_label(bars)
```

```python
cui = df.groupby(['YOB'],as_index =False)['Smoker'].sum().sort_values(by = 'Smoker',ascending = False)
sns.barplot(data = cui,x = 'YOB',y = 'Smoker',palette = 'deep')
```

### from above graph we conclude that their is maximum number of people who smoke socially in a year

### Area Code

```python
p = df.pivot(index ='ID',columns ='YOB',values ='Area code')
sns.heatmap(p)
```

```python
sns.lmplot(x = 'ID', y  = 'Area code', hue = 'Service Rating', data = df)
```

```python
df.columns
```

### Budget

```python
v = sns.regplot(x = 'Budget', y = 'YOB', data = df)
for bars in v.containers:
     v.bar_label(bars)
```

```python
cui = df.groupby(['Gender'],as_index =False)['Budget'].sum().sort_values(by = 'Budget',ascending = False)
sns.barplot(data = cui,x = 'Gender',y = 'Budget')
```

### from above graph we conclude that their is maximum number of male has more budget and number of females has least budget.

### Cuisines

```python
ax = sns.countplot(x = 'Cuisines', data = df, palette = 'colorblind')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
cui = df.groupby(['Cuisines'],as_index =False)['Food Rating'].sum().sort_values(by = 'Food Rating',ascending = False)
sns.barplot(data = cui,x = 'Cuisines',y = 'Food Rating')
```

### Here it shows that Japanese food has maximum rating and Italian has minimum.

### Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital Status')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
cui = df.groupby(['Marital Status','Gender'],as_index =False)['Food Rating'].sum().sort_values(by = 'Food Rating',ascending = False)
sns.barplot(data = cui,x = 'Marital Status',y = 'Food Rating',hue ='Gender')
```

### Here we conclude that their is maximum number of single male and least number of divorced females

### Activity

```python
ax = sns.countplot(data = df, x = 'Activity',palette ='dark')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
cui = df.groupby(['Activity'],as_index =False)['Food Rating'].sum().sort_values(by = 'Food Rating',ascending = False)
sns.barplot(data = cui,x = 'Activity',y = 'Food Rating',palette ='dark')
```

### Here we conclude that their is maximum number of students rated food than professionals
